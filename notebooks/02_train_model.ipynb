```python
# Dashboard GIF Generator
# Create an animated GIF from the dashboard screenshots.
```

```python
from PIL import Image
from pathlib import Path

image_dir = Path("../images")

image_files = [
    image_dir / "overview.png",
    image_dir / "ops.png",
    image_dir / "ml.png",
    image_dir / "strategy.png",
    image_dir / "deep-dive.png",
]

frames = [Image.open(img) for img in image_files]

# Resize all frames to match the first image's dimensions
target_size = frames[0].size
frames = [f.resize(target_size, Image.LANCZOS) for f in frames]

output_path = image_dir / "dashboard_view.gif"
frames[0].save(
    output_path,
    save_all=True,
    append_images=frames[1:],
    duration=2000,
    loop=0,
)

print(f"GIF saved to {output_path} ({output_path.stat().st_size / 1024:.0f} KB)")
```
